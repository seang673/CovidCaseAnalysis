### **Quality Issue 1: FILL IN**

### **Quality Issue 2: FILL IN**

```python
#FILL IN - Apply the cleaning strategy
chi_df_update = chi_df[['Date', 'Cases - Total', 'Hospitalizations - Total', 'Deaths - Total' ]]
```

```python
#FILL IN - Validate the cleaning was successful
chi_df_update.head()
```

Justification: The relevant columns in the Chicago dataframe are now being displayed

### **Tidiness Issue 1: FILL IN**

```python
#FILL IN - Apply the cleaning strategy
nyc_df_update.info()
```

```python
#change date format and narrow the timeframe to summer 2020 only
nyc_df_update['date_of_interest'] = pd.to_datetime(nyc_df_update['date_of_interest'], format='%m/%d/%Y')
nyc_df_update = nyc_df_update.loc[(nyc_df_update['date_of_interest'] >= '2020-06-01')
                     & (nyc_df_update['date_of_interest'] < '2020-08-22')]
nyc_df_update = nyc_df_update.reset_index(drop=True)
```

```python
#FILL IN - Validate the cleaning was successful
nyc_df_update.head(10)
```

```python
nyc_df_update.tail(5)
```

Justification: The relevant dates are displayed

### **Tidiness Issue 2: FILL IN**

```python
#FILL IN - Apply the cleaning strategy
#Drop row with missing date
chi_df_update = chi_df_update.dropna(subset=['Date'])
```

```python
#Fill in NA values for hospitalization with the mean value
chi_df_update['Hospitalizations - Total'] = chi_df_update['Hospitalizations - Total'].fillna(chi_df_update['Hospitalizations - Total'].mean())
```

```python
#Convert 'Date' column to datetime and sort by date
chi_df_update['Date'] = pd.to_datetime(chi_df_update['Date'], format='%m/%d/%Y')
chi_df_update = chi_df_update.sort_values(by='Date', ignore_index=True)
```

```python
#Change column names
new_column_names = ['Date', 'total_cases', 'total_hospitalizations', 'total_deaths']
chi_df_update.columns = new_column_names
```

```python
#Only keep dates that are in timeframe of summer 2020
chi_df_update = chi_df_update.loc[(chi_df_update['Date'] >= '2020-06-01')
                     & (chi_df_update['Date'] < '2020-08-22')]
chi_df_update = chi_df_update.reset_index(drop=True)
```

```python
#FILL IN - Validate the cleaning was successful
chi_df_update.head(10)
```

```python
chi_df_update.tail()
```

Justification: The columns with the relevant variables and rows with relevant dates are displayed

### **Combine datasets into one**

Depending on the datasets, you can also peform the combination before the cleaning steps.

```python
#FILL IN - Remove date column from the chicago dataframe, as we prepare to combine the two datasets
chi_df_update = chi_df_update.drop(['Date'], axis=1)
chi_df_update.head()
```

```python
#combine both dataframes into one
df_merged = pd.concat([nyc_df_update, chi_df_update], ignore_index=True, sort=False)
```

```python
#make column variables distinct for each city
new_column_names = ['Date', 'nyc_case_count', 'nyc_hospitalization_count', 'nyc_death_count', 'chi_case_count', 'chi_hospitalization_count', 'chi_death_count']
df_merged.columns = new_column_names
```

```python
df_merged.reset_index(drop=True, inplace=True)
```

```python
df_merged.describe()
```

## 5. Answer the research question

### **5.1:** Define and answer the research question 
Going back to the problem statement in step 1, use the cleaned data to answer the question you raised. Produce **at least** two visualizations using the cleaned data and explain how they help you answer the question.

*Research question:* During the summer of 2020, was there a significant statistical difference between average cases and deaths of covid-19 patients between NYC and Chicago? 

```python
#Visual 1 - FILL IN
import matplotlib.pyplot as plt

ax = plt.axes()
ax.bar(df_merged.index, df_merged['nyc_case_count'], width=0.5, color='g', label='nyc_case_count')
ax.bar(df_merged.index, df_merged['chi_case_count'], width=0.5, color='r', label='chi_case_count')
ax.get_xaxis().set_visible(False)
plt.legend()
plt.show()
```

*Answer to research question:* NYC did have more covid-19 cases over summer of 2020, likely due to it being a larger city, however it was surprisingly not a significantly larger amount 

```python
#Visual 2 - FILL IN
import matplotlib.pyplot as plt
ax = plt.axes()
ax.bar(df_merged.index, df_merged['nyc_death_count'], width=0.5, color='g', label='nyc_death_count')
ax.bar(df_merged.index, df_merged['chi_death_count'], width=0.5, color='r', label='chi_death_count')
ax.get_xaxis().set_visible(False)
plt.legend()
plt.show()
```

*Answer to research question:* Relative to the case counts, NYC does appear to have a significantly larger covid-19 related death count over summer 2020 than Chicago.
